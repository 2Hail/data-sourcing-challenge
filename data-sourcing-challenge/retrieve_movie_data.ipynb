### Access the New York Times API

```python
# Set the base URL
url = "https://api.nytimes.com/svc/search/v2/articlesearch.json?"

# Filter for movie reviews with "love" in the headline
# section_name should be "Movies"
# type_of_material should be "Review"
filter_query = 'section_name:"Movies" AND type_of_material:"Review" AND headline:"love"'

# Use a sort filter, sort by newest
sort = "newest"

# Select the following fields to return:
# headline, web_url, snippet, source, keywords, pub_date, byline, word_count
field_list = "headline,web_url,snippet,source,keywords,pub_date,byline,word_count"

# Search for reviews published between a begin and end date
begin_date = "20130101"
end_date = "20230531"

# Build URL
query_url = f"{url}api-key={nyt_api_key}&fq={filter_query}&sort={sort}&fl={field_list}&begin_date={begin_date}&end_date={end_date}"
query_url
```

```python
reviews_list = []

for page in range(20):
    query_url = f"{url}api-key={nyt_api_key}&fq={filter_query}&sort={sort}&fl={field_list}&begin_date={begin_date}&end_date={end_date}&page={page}"
    response = requests.get(query_url)

    if response.status_code == 200:
        reviews = response.json()
        response_list = reviews["response"]["docs"]
        time.sleep(12)

        try:
            for doc in reviews["response"]["docs"]:
                reviews_list.append(doc)
            print(f"Checked Page {page}")
        except KeyError:
            print(f"No Results: {page}")
            break
    else:
        print(f"Failed to Retrieve: {page}")
        break
print(f"Total Reviews: {len(reviews_list)}")
```

### Access The Movie Database API

```python
# Prepare The Movie Database query
url = "https://api.themoviedb.org/3/search/movie?query="
tmdb_key_string = "&api_key=" + tmdb_api_key
```

```python
tmdb_movies_list = []
request_counter = 1
for title in title_list:
    if request_counter % 50 == 0 and request_counter != 0:
        time.sleep(10)
        try:
            response = requests.get(f"{url}{title}{tmdb_key_string}")
            response_json = response.json()
            if response_json.get("results"):
                movie_id = response.json()["results"][0]["id"]
                movie_details_response = requests.get(f"https://api.themoviedb.org/3/movie/{movie_id}?{tmdb_key_string}")
                movie_details = movie_details_response.json()
                genres = [genres["name"] for genre in movie_details["genres"]]
                spoken_languages = [language["english_name"] for language in movie_details["spoken_languages"]]
                production_countries = [country["name"] for country in movie_details["production_countries"]]
                tmdb_movies_list.append({
                    "title": title,
                    "genres": genres,
                    "spoken_languages": spoken_languages,
                    "production_countries": production_countries
                })
                print(f"Title Found: {title}")
            else:
                print(f"Title Not Found {title}")
        except Exception as e:
            print(f"Failed Processing{title}: {e}")
    request_counter += 1
        
print(f"Total Movies Found: {len(tmdb_movies_list)}")
```
